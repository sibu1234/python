```python
#with repetation
def union(list1,list2):
    final_list=list1 + list2
    return final_list
list1=[4,5,9,6,3,12,21,17]
list2=[8,7,2,10,11,15,13,16]
print(union(list1,list2))
```

```python
#with repetation and ordered
def union(list1,list2):
    final_list=sorted(list1 + list2)
    return final_list
list1=[4,5,9,6,3,12,21,17]
list2=[8,7,2,10,11,15,13,16]
print(union(list1,list2))
```

```python
#without repetation
def union(list1,list2):
    final_list=list(set(list1)|set(list2))
    return final_list
list1=[2,6,4,8,7,5,3,1,5,3,7,9,4]
list2=[7,5,2,6,3,4,8]
print(union(list1,list2))
```

```python
#without repeatation more than 1 list
def union(list1,list2,list3):
    final_list=list(set().union(list1,list2,list3))
    return final_list
list1=[1,2,3,4]
list2=[3,4,5,6]
list3=[5,6,7,8]
print(union(list1,list2,list3))
```

```python
def intersection(list1,list2):
    return list(set(list1) & set(list2))
list1=[2,3,4,5,6]
list2=[4,5,6,7,8]
print(intersection(list1,list2))
```
