```python
def plot_iris(iris,col1,col2):
    import seaborn as sns
    import matplotlib.pyplot as plt
    
    sns.lmplot(x=col1, y=col2,
              data=iris,
              hue="Species",
              fit_reg=False)
    plt.xlabel(col1)
    plt.ylabel(col2)
    
    plt.title("Iris Species")
    plt.show()
```

```python
plot_iris(iris.data,"Petal_Width","Sepal_Length")
plot_iris(iris.data,"Sepal_Width","Sepal_Length")
```

```python
plot_iris(iris.data,"Sepal_Length","Sepal_Width")
plot_iris(iris.data,"Sepal_Length","Petal_Length")
plot_iris(iris.data,"Sepal_Length","Petal_Width")
plot_iris(iris.data,"Sepal_Width","Petal_Length")
plot_iris(iris.data,"Sepal_Width","Petal_Width")
plot_iris(iris.data,"Petal_Length","Petal_Width")
```

```python
from sklearn.preprocessing import scale
import pandas as pd
```

```python
colnames=["Sepal_Length","Sepal_Width","Petal_Length","Petal_Width"]
```

```python
iris_scaled=scale(iris.data[colnames])
iris_scaled
```

```python
iris_scaled=pd.DataFrame(iris_scaled,columns=colnames)
print(iris_scaled)
```

```python
print(iris_scaled.describe().round())
```

```python
levels={"setosa":0,"versicolor":1,"virginica":2}
levels["versicolor"]
```

```python
iris_scaled["Species"]=[levels[x] for x in iris.data["Species"]]
```

```python
iris_scaled.tail()
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
np.random.seed(123)
```

```python
m=np.asmatrix(iris_scaled)
m
# here m is numpy array i.e list of lists. Array itself is a matrix without heading.
```

```python
iris_split=train_test_split(m,test_size=75)
```

```python
iris_split
```

```python
x_train=iris_split[0][:,:4]
y_train=np.ravel(iris_split[0][:,4])
#ravel is used for 1D array
```

```python
x_test=iris_split[1][:,:4]
y_test=np.ravel(iris_split[1][:,4])
#Flattern is used for 2D array
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

# Training and Evaluation of Machine Learning model using KNN Algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN_Mod=KNeighborsClassifier(n_neighbors=3)
```

```python
KNN_Mod.fit(x_train,y_train)
```

```python
iris_test=pd.DataFrame(x_test,columns=colnames)
iris_test
```

```python
iris_test["predicted"]=KNN_Mod.predict(x_test)
iris_test
```

```python
iris_test["correct"]=[1 if x==z else 0 for x,z in zip(iris_test["predicted"],y_test)]
```

```python
iris_test.sample(10)
```

```python
accuracy=100*(sum(iris_test["correct"])/iris_test.shape[0])
```

```python
print(accuracy)
```
